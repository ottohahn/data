### Trends vs Leads

The below data was taken directly from the admin panel on 7/19/16.

```python
Image(filename = 'images/thumbtack_admin.png')
```

```python
positions = ['Android Engineer',
             'Data Scientist',
             'Director of IT',
             'Front End Engineer',
             'Head of Data Science',
             'iOS Engineer',
             'Security Engineer',
             'Site Reliability Engineer',
             'Software Engineer',
             'Engineering Director']
leads = [4870, 14348, 2720, 2720, 39418, 767, 4961, 100, 4212, 39418]
trends = [4, 1, 3, 5, 0, 3, 3, 6, 2, 0]
```

```python
plt.scatter(trends, leads)
z = np.polyfit(trends, leads, 1)
p = np.poly1d(z)
plt.plot(trends, p(trends), "r")
plt.title("Trends vs Leads")
plt.xlabel("Trends")
plt.ylabel("Leads")
```

When briefly looking at leads versus trends, it's immediately noticeable that jobs with more trends tend to have fewer leads. By increasing trends for each position, we should hopefully see an immediate decrease in the number of lead matches.
